```python

#Opens the image
from PIL import Image
def cropIm(fldrIn, fn, fldrOut, type='png'):
    path = fldrIn + fn
    try:
        im = Image.open(path)
    except:
        im = Image.open(path.replace('PNG', 'png')   ) 
    rgb_im = im.convert('RGB')

    left = 0
    top = 0
    right = left + 788
    bottom = top + 1024

    #Crops the image
    rgb_im = rgb_im.crop((left, top, right, bottom))
    
    # resize - otherwise renet cant allocate enough memory
    MAX_SIZE = (788, 1024)
    rgb_im.thumbnail(MAX_SIZE)

    print(im.size, fn)
    #Saves the image
    rgb_im.save(fldrOut + fn[:-4] + f'.{type}', format=type)
    rgb_im.close()
# cropIm(fldrIn, fn, fldrOut)

import glob
FireList = []
NoFireList = []

fldrIn = "data/orig/train/Fire/"
fldrOut = "data/output/train/Fire/"
for fire, fn in enumerate(glob.glob(fldrIn+"*.PNG")):
    fn_trunc = fn.split('/')[-1]
    FireList.append(fn_trunc)
    cropIm(fldrIn, fn_trunc, fldrOut, 'png')
    
print('Count fire files', fire)

fldrIn = "data/orig/train/NoFire/"
fldrOut = "data/output/train/NoFire/"
for noFire, fn in enumerate(glob.glob(fldrIn+"*.PNG")):
    fn_trunc = fn.split('/')[-1]
    NoFireList.append(fn_trunc)
    cropIm(fldrIn, fn_trunc, fldrOut, 'png')
print('Count NoFire files', noFire)

FireList = []
NoFireList = []

fldrIn = "data/orig/val/Fire/"
fldrOut = "data/output/val/Fire/"
for fire, fn in enumerate(glob.glob(fldrIn+"*.PNG")):
    fn_trunc = fn.split('/')[-1]
    FireList.append(fn_trunc)
    cropIm(fldrIn, fn_trunc, fldrOut, 'png')
    
print('Count fire files', fire)

fldrIn = "data/orig/val/NoFire/"
fldrOut = "data/output/val/NoFire/"
for noFire, fn in enumerate(glob.glob(fldrIn+"*.PNG")):
    fn_trunc = fn.split('/')[-1]
    NoFireList.append(fn_trunc)
    cropIm(fldrIn, fn_trunc, fldrOut, 'png')
print('Count NoFire files', noFire)
```
